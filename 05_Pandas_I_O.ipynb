## Importa√ß√µes

**create_engine**, que serve para criar o motor, ou seja, a conex√£o com o banco de dados que o far√° funcionar.

**MetaData** Essa √© uma classe que representa os meta dados do banco de dados, isso inclui informa√ß√µes sobre as tabelas, colunas e outros objetos.

**Table** que representa uma tabela em um banco de dados e √© usada para manipular dados em uma tabela espec√≠fica.

**inspect** que permite inspecionar a estrutura do banco de dados. Em seguida, executamos a c√©lula.

````python
engine = create_engine('sqlite:///:memory:')
````

###**Lendo uma consulta SQL**

````python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
````

Dentro das aspas escrevemos exatamente o que queremos consultar em SQL. Nesse caso, o que o time de dados quer √© separar as pessoas que est√£o com a categoria de renda preenchidas como "Empregado", ou seja, s√≥ as pessoas que est√£o trabalhando no momento.

Quando trabalhamos com SQL utilizamos cl√°usulas, ou seja, palavras que possuem fun√ß√µes espec√≠ficas para a consulta.

Lembrando que em SQL as cl√°usulas s√£o escritas em caixa alta.

Sendo assim, ap√≥s as aspas, escrevemos SELECT seguido de *. Esse asterisco indica que queremos selecionar todas as colunas da tabela.

Agora, precisamos especificar qual tabela √© essa. Ent√£o, ap√≥s * inserimos a cl√°usula FROM clientes.

Na mesma linha, escrevemos a cl√°usula WHERE, que filtra as informa√ß√µes, seguido do nome da coluna que queremos consultar Categoria_de_renda. Logo ap√≥s escrevemos ="Empregado" e executamos.

Estamos pedindo para selecionar todas as colunas da tabela clientes na qual a categoria de renda seja igual a "Empregado".

````python
empregados = pd.read_sql(query, engine) #le uma consulta
````

````python
empregados
````

apenas empregados

````python
empregados.to_sql('empregados', con=engine, index=False)
````

````python
pd.read_sql_table('empregados', engine) #le uma tabela inteira
````

````python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade',
                                                 'Rendimento_anual'])
````

Lembre-se sempre dessa diferen√ßa, o read_sql permite ler uma consulta e o read_sql_table uma tabela completa.

Atualizando um banco de dados

Suponhamos que a pessoa de ID 5008804 cancelou sua conta na institui√ß√£o financeira. Isso significa que ela n√£o faz mais parte do quadro de clientes.

Para tir√°-la da tabela precisamos realizar uma consulta. Ent√£o, escrevemos query = ''. Dentro das aspas, inserimos a instru√ß√£o DELETEseguido da cl√°usula FROM clientes. Agora, especificamos qual cliente ser√° deletado, ent√£o, escrevemos WHERE seguido do nome da coluna ID_Cliente e o n√∫mero do ID 5008804.

Apertamos "Enter" para a linha de baixo. Agora, fazemos um contexto with. Isso garante que a conex√£o com o banco de dados seja fechada corretamente ap√≥s a execu√ß√£o da consulta. Depois, usamos o m√©todo engine.connect(), seguido de as conn.

````python
query = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
with engine.connect() as conn:
    conn.execute(query)
````

Com isso estamos usando um m√©todo chamado connect para estabelecer uma conex√£o com o banco de dados. Estamos pedindo para executar a query com o m√©todo execute que quando chamado, executa a instru√ß√£o SQL diretamente no banco de dados, usando a conex√£o estabelecida.

````python
pd.read_sql_table('clientes', engine)
````

````python
query = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'
with engine.connect() as conn:
    conn.execute(query)
````
