# Calculando o coeficiente angular (a)

image.png

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
(X**2).shape
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
```

# Calculando o coeficiente linear (b)

image.png

```python
b = np.mean(Y) - a*np.mean(X)
```

```python
y = a*X+b
```

```python
np.linalg.norm(Moscow-y) #ajuste melhor
```

# Regressão


```python
plt.plot(datas,Moscow)
plt.plot(x,y)
plt.plot(41.5,41.5*a+b,'*r') #pontinho no gráfico
```

random.seed() seleciona um número semente fixo para inicializar o gerador aleatório, optaremos por 16.

```python
np.random.seed(16)
np.random.uniform(low=0.10, high=0.90,size=100) #usamos seguido da função seed()
```

```python
np.random.uniform(low=0.10, high=0.90,size=100) #perceba que gerou números aleatórios novamente
```

```python
np.random.seed(84)
coef_angulares = np.random.uniform(low=0.10,high=0.90,size=100)
norma = np.array([])
for i in range(100):
    norma = np.append(norma, np.linalg.norm(Moscow-(coef_angulares[i]*X+b)))
```

se o valor passado para o seed é importante, e a resposta é SIM! Na verdade, serão geradas sequências diferentes, mas isso não interfere na qualidade dos números aleatórios gerados. Se digitarmos um valor muito alto, possivelmente teremos um erro, mas se optarmos por um valor de um intervalo possível, podemos escolher qualquer número que a qualidade não sofrerá interferência.
