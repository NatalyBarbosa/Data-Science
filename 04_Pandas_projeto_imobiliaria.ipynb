# Método mean()

O Pandas possui um método chamado mean() que nos permite calcular a média de um conjunto de dados. Vamos começar calculando a média dos valores da coluna "Valor". Passaremos dados, selecionaremos a coluna 'Valor', e adicionaremos .mean():

```python
dados['Valor'].mean()
```

# Método groupby()

O método groupby() funciona em conjunto com outras funções do Pandas, como a sum de soma ou a mean que é a média. Com essas funções, podemos fazer agregações conforme o agrupamento que realizamos.

```python
dados.groupby('Tipo').mean() #função que agrupa
```

Se você estiver usando a versão 1 da biblioteca Pandas, automaticamente esse método de agregação será aplicado unicamente nas colunas numéricas. No entanto, se você estiver utilizando da versão 2 em diante, esse comportamento não será o mesmo. Nesse caso, ao fazer o agrupamento e aplicar o método de agregação, ele será aplicado a todas as colunas, incluindo as de texto. Isso pode gerar um erro.

```python
dados.groupby('Tipo').mean(numeric_only=True)
```

Para evitar que isso aconteça, podemos especificar um parâmetro no método mean() (e nos outros métodos de agregação também): numeric_only=True. Com isso, especificamos que a média será calculada apenas nas colunas numéricas e não teremos nenhum erro. Como foi mostrado acima.

# Melhorando a visualização do dataframe


Selecionando uma coluna numérica específica antes da aplicação do método mean(). Isso significa que esse método não será mais aplicado a todas as colunas do dataframe, mas apenas na coluna selecionada. Então, não precisamos mais utilizar o parâmetro numeric_only=True no método. Podemos apagá-lo, ainda que deixá-lo não vá gerar nenhum erro.


```python
dados.groupby('Tipo')['Valor'].mean() #o valor médio do alguel de cada tipo de imóvel
```

```python
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor') #melhorando a visualização
```

# Método to_frame()

Vamos copiar o código com value_counts(). Desta vez, utilizaremos o método to_frame() que converte uma series em um dataframe.

```python
df.Tipo.value_counts(normalize=True).to_frame()
```

Observação: Na nova atualização da biblioteca Pandas, a coluna de valores proporcionais, quando no dataframe, será nomeada automaticamente como proportions. Com essa mudança, é necessário alterar também a coluna selecionada para a criação da variável df_percentual tipo:


```python
df['Tipo'].value_counts(normalize=True).to_frame().sort_values('proportion') #menor para o mair
```

```python
df_percentual_tipo = df['Tipo'].value_counts(normalize=True).to_frame().sort_values('proportion')
```

Passamos df_percentual_tipo, nome da variável onde salvamos o dataframe.

Depois, .plot(), para plotarmos o gráfico. Nos parênteses, indicamos que estamos plotando um gráfico de barras verticais.

Com figsize definimos o tamanho, com color a cor das barras (verde).

Ainda nos parênteses do .plot(), passamos dois parâmetros: xlabel, que é o nome do eixo x, e o y_label, que é o nome do eixo y.

Vamos executar o código e visualizar o gráfico!

```python
df_percentual_tipo = df['Tipo'].value_counts(normalize=True).to_frame().sort_values('Tipo')

df_percentual_tipo.plot(kind='bar', figsize=(14, 10), color ='green', edgecolor='black',
                        xlabel = 'Tipos', ylabel = 'Percentual');
```

O retorno é um gráfico de barras, onde o eixo x corresponde aos tipos únicos que temos na nossa base de dados (da origem para a extremidade positiva): studio; loft; casa de vila; flat; quitinete; casa; casa de condomínio; aparamento.

O eixo y corresponde ao percentual dos tipos da nossa base de dados. Ele está graduado de 0.0 a 0.8, com intervalos de 01.

Analisando o gráfico, podemos notar que os imóveis do tipo apartamento representam a maioria da nossa base de dados: mais de 80% dos dados. Pensando nisso:

# Criando colunas numéricas

Começaremos importando novamente nossa base de dados. Isso porque, para o site da imobiliária, as informações de todos os imóveis são importantes. Mas lembre-se que, anteriormente, trabalhamos apenas com apartamentos, então não faz sentido continuar utilizando o mesmo dataframe.

Faremos a importação salvando-a em uma variável url. Em seguida, fazemos a leitura na variável dados:

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
#criação de uma coluna: valor por mês, a soma de valor e condomínio

dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
#criação de uma coluna: valor por ano

dados['Valor_por_ano'] = dados['Valor_por_mes']*12 + dados['IPTU']
dados.head()
```

# Criando uma coluna categórica

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] #concatenando strings
dados.head()
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'] + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'] + ' vaga(s) de garagem.'
```

Erro: Concatenamos strings e não inteiros! Devemos convertê-los em str usando o .astype(str)

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

# Criando uma coluna binária

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

Neste comando, usamos o método .apply(), utilizado para aplicar uma determinada função em um conjunto específico de dados. No nosso caso, esse conjunto é a coluna "Suites", e a função, é do tipo lambda - funções personalizadas que podem ser escritas em uma única linha.

Nesta função, x equivale a cada linha da coluna de suítes. Então, fizemos uma condicional if em que se o valor de x, ou seja, o registro da linha, for maior que zero, deve ser colocado "Sim", caso contrário (else), coloca-se "Não". "suites" no caso.
